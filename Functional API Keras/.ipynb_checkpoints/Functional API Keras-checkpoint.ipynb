```python
# lets differenciate between the features and the target variable
features = df.drop('target', axis = 1)
target = df[['target']]
features.head()
```

```python
target.head()
```

```python
# now lets differenciate between the categorical values and the numerical values 
categorical_features = features[['sex', 'fbs','exang','cp','ca','slope','thal','restecg']].copy()

categorical_features.head()
```

```python
numerical_features = features[['age','trestbps','chol','thalach','oldpeak']].copy()

numerical_features.head()
```

```python
# if you look at the mean and standard deviation have very different values
# machine learning model especially neural networks tend to be far more robust
# when they are trained using numeric features that has the same scale.
# The way we achieve this in machine learning is processing the numeric 
# features using a technique called standardisation.The standardisation technique
# is performed using standard scaler in scikit learn. standardisation is a column 
# wise operation where for every value in a column we substract mean and divide 
# by the standard deviation.This expresses all of our data in terms of standard
# deviation or Z-scores. Here we use the standard scaler
standardScaler =  StandardScaler()

numeric_features = pd.DataFrame(standardScaler.fit_transform(numerical_features),
                               columns = numerical_features.columns,
                               index = numerical_features.index)

numeric_features.describe().T
```

```python
# now lets join all the feature. That is we will combine 
# numeric_features and the categorical features
processed_features = pd.concat([numeric_features, categorical_features], axis = 1,
                              sort = False)

processed_features.head()
```

```python
# lets now split the training and testing data set 
x_train, x_test, y_train, y_test = train_test_split(processed_features,
                                                  target,
                                                  test_size = 0.2,
                                                  random_state = 1)
```

```python
# now lets spilt the training data into training data and validation data
x_train, x_val,y_train,y_val = train_test_split(x_train,
                                               y_train,
                                               test_size = 0.15,
                                               random_state = 10)
```

```python
# check the number of records for features
x_train.shape, x_val.shape, x_test.shape
```

```python
# check the number od records for target
y_train.shape, y_val.shape, y_test.shape
```

```python
# we are now ready to build the model as our preprocessing work 
# completed using the functional API 
def build_model():
    inputs = tf.keras.Input(shape = (x_train.shape[1],)) # input to my model
    
    dense_layer1 = layers.Dense(12, activation = 'relu')# first layer is a dense layer with 12 neurons
    x = dense_layer1(inputs) # pass the input layer into the first dense layer by invoking dense_layer1
    
    # here all the layers are callable functions 
    dropout_layer = layers.Dropout(0.3) # next layer is the dropout layer
    x = dropout_layer(x)
    
    dense_layer2 = layers.Dense(8, activation = 'relu')
    x = dense_layer2(x)
    
    predictions_layer = layers.Dense(1, activation = 'sigmoid') # this is the last prediction layer with sigmoid activation function
    prediction = predictions_layer(x)
    
    model = tf.keras.Model(inputs = inputs, outputs = prediction)
    
    model.summary()
    
    model.compile(optimizer = tf.keras.optimizers.Adam(0.001),
                 loss = tf.keras.losses.BinaryCrossentropy(),
                 metrics = ['accuracy', tf.keras.metrics.Precision(0.5),
                           tf.keras.metrics.Recall(0.5),])
    
    return model
```

```python
# lets go ahead and create the model 
model = build_model()
```

```python
# we can also view the layers in the neural network
keras.utils.plot_model(model, show_shapes = True)
```

```python
# now we build the pipe lines to transform your data and feed data into ML model 
# for training, evluation and prediction 
dataset_train = tf.data.Dataset.from_tensor_slices((x_train.values, y_train.values))
dataset_train = dataset_train.batch(16)

dataset_train.shuffle(126)
```

```python
num_epochs = 100
```

```python
# same for validation data set 
dataset_val = tf.data.Dataset.from_tensor_slices((x_val.values,y_val.values))
dataset_val = dataset_val.batch(16)
```

```python
# now we will train the model 
training_history = model.fit(dataset_train, epochs = num_epochs, validation_data = dataset_val)
```

```python
# we will take a look at the metrics that has been recorded during the training
training_history.history.keys()
```

```python
# will use the malplotlib to display training related details 
# plot the accracy, loss precision and recall on the training data
train_acc = training_history.history['accuracy']
train_loss = training_history.history['loss']

precision = training_history.history['precision']
recall = training_history.history['recall']

epoch_range = range(num_epochs)

plt.figure(figsize = (14,8))

plt.subplot(1, 2, 1)

plt.plot(epoch_range,train_acc, label = 'Training Accuracy')
plt.plot(epoch_range, train_loss, label = 'Training Loss')

plt.title('Accuracy and Loss')
plt.legend()

plt.subplot(1, 2, 2)

plt.plot(epoch_range, precision, label = 'Precision')
plt.plot(epoch_range, recall, label = 'Recall')

plt.title('Precision and Recall')
plt.show()
```

```python
# now we will evulate the model on the test data
score = model.evaluate(x_test, y_test)

score_df = pd.Series(score, index = model.metrics_names)
score_df
```

```python
# now we will ge the predicted values for the model
y_pred = model.predict(x_test)
y_pred[:10]
```

```python
# y_pred is the proability of the patients who has the heart diseases
# we will have a threshold of 0.5
y_pred = np.where(y_pred >= 0.5 , 1 , y_pred)
y_pred = np.where(y_pred < 0.5, 0, y_pred)
y_pred[:10]
```

```python
# from the actuall data and the predicted data we will compare as 
# a data frame side by side 
pred_results = pd.DataFrame({'y_test': y_test.values.flatten(),
                            'y_pred': y_pred.flatten().astype('int32')}, index = range(len(y_pred)))
```

```python
pred_results.head(10)
```

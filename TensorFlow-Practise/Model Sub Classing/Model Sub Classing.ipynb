```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.utils import to_categorical
```

```python
# the data set here we use it the wine data set from the sklearn
wine_data = datasets.load_wine()
```

```python
print(wine_data['DESCR'])
```

```python
# will set the data in the pandas data frame 
data = pd.DataFrame(data = wine_data['data'], columns = wine_data['feature_names'])
data['target'] = wine_data['target']
data.head(10)
```

```python
# lets now look at the shape of the data frame 
data.shape
```

```python
# now lets check the statistics of the data frame
data.describe().T

# here we can see that all features are numeric in nature and the mean 
# and the standard deviation have the different scales. we need to standardise the values
```

```python
# check the data set with any null values
data.isna().sum()
```

```python
# lets explore the data a bit .lets take a look at the target variable
data['target'].value_counts()
```

```python
# now we are curious about the alcohol content is distributed 
# across the wines in our data set
sns.distplot(data['alcohol'], kde = 1)

# the majority of the alcohol are between 12 and 14 percent 
```

```python
# also we will check how the alcohol content vary based on the different 
# category of wine . a box plot
plt.figure(figsize = (10,6))

sns.boxplot('target', 'alcohol', data = data)

plt.xlabel('class', fontsize = 20)
plt.ylabel('alcohol',fontsize = 20)

plt.show()

# the class of 0 has the highest content of alcohol, class 1 has the lowest 
# alcohol and the class 2 has the medium alcohol content
```

```python
# lets now create the features and the target in our ML model
features = data.drop('target', axis = 1)

target =  data[['target']]
```

```python
# lets check the features columns 
features.columns
```

```python
target.head()
```

```python
# here the claasifier is the muti class classfier.The wine is of type 
# 3 types of catgories. target needs to encoded in one hot encoded
# form. to convert the target to one hot encoding we use to_categorical method
target = to_categorical(target, 3)
target
```

```python
#now lets standise the values in the data frame 
standardscaler = StandardScaler()

processed_features = pd.DataFrame(standardscaler.fit_transform(features),
                                 columns = features.columns,
                                 index = features.index)
processed_features.describe().T
```

```python
# now we will spilt the data frame into train anf test
x_train, x_test, y_train,y_test = train_test_split(processed_features,
                                                 target,
                                                 test_size = 0.2,
                                                 random_state =1)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
# lets build the network

class WineClassificationModel(Model):
    
    def __init__(self, input_shape):
        
        super(WineClassificationModel, self).__init__()
        
        self.d1 = layers.Dense(128, activation = 'relu', input_shape = [input_shape])
        self.d2 = layers.Dense(64, activation = 'relu')
        
        self.d3 = layers.Dense(3, activation = 'softmax')
        
    def call(self, x):
        x = self.d1(x)
        x = self.d2(x)
        x = self.d3(x)
        return x       
```

```python
# we will now instantiate wineclassificationmodel and pass in the
# shape of the input data
model = WineClassificationModel(x_train.shape[1])

model.compile(optimizer = keras.optimizers.SGD(lr = 0.001),
             loss = tf.keras.losses.CategoricalCrossentropy(),
             metrics = ['accuracy'])
```

```python
num_epochs = 500
```

```python
# wil start the training process
training_history = model.fit(x_train.values,
                            y_train,
                            validation_split = 0.2,
                            epochs = num_epochs,
                            batch_size = 48)
```

```python
# we will take a look at the metrics that has been recorded during the training
training_history.history.keys()
```

```python
# will use the malplotlib to display training related details 
# plot the accracy, loss validation accuracy  and validation loss on the training data
train_acc = training_history.history['accuracy']
train_loss = training_history.history['loss']

precision = training_history.history['val_accuracy']
recall = training_history.history['val_loss']

epoch_range = range(num_epochs)

plt.figure(figsize = (14,8))

plt.subplot(1, 2, 1)

plt.plot(epoch_range,train_acc, label = 'Training Accuracy')
plt.plot(epoch_range, train_loss, label = 'Training Loss')

plt.title('Training')
plt.legend()

plt.subplot(1, 2, 2)

plt.plot(epoch_range, precision, label = 'validation Accuracy')
plt.plot(epoch_range, recall, label = 'validation Loss')

plt.title('validation ')
plt.show()
```

```python

# now we will evulate the model on the test data
score = model.evaluate(x_test, y_test)

score_df = pd.Series(score, index = model.metrics_names)
score_df
```

```python
# now we will ge the predicted values for the model
y_pred = model.predict(x_test)
y_pred[:10]
```

```python
# y_pred is the proability of the class of wine 
# we will have a threshold of 0.5
y_pred = np.where(y_pred >= 0.5 , 1 , y_pred)
y_pred = np.where(y_pred < 0.5, 0, y_pred)
y_pred[:10]
```

```python
#now we will calculate the accuracy score 
accuracy_score(y_test, y_pred)
```

```python
from random import randint

import cv2
import os

import pandas as pd
import numpy as np
 
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.preprocessing import image

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# lets go ahead and load the data set from tensorflow data set 
# cifar 10 is a labeled subset of the 80million tiny images data sets
# This contains a total of 60k images in 10 classes
# source https://www.cs.toronto.edu/~kriz/cifar.html
cifer_10 = tf.keras.datasets.cifar10

(train_images,train_labels), (test_images,test_labels) = cifer_10.load_data()
```

```python
# now lets look the number of images and the pixels size of the images 
train_images.shape, test_images.shape

# here there are 50k images for the training with 32 * 32 pixels and 3 channel 
# (rgb)
# for the test data we have the test data of 10k images 
```

```python
# lets access the first images to checkout the pixel values 
train_images[0]

# here we have the 3 channel i.e., for the rgb. The pixel value is from 0 to 255
```

```python
# we will now take a look at the labels associated in the each image in out training
train_labels[:20]

# here the label are the numeric value from 0 to 9 representing 10 classes 
```

```python
# each numberic item corresponds to dictionary. This is the 10 lookup categories
# represented in the cifar 10 data set 
lookup = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]
```

```python
# In order to view and explore the images present in the cifar 10 data set
# will set up a helper function show_img which takesin images, label, number of 
# images selected at random 
def show_img(images, labels, n_images):
    random_int = randint(0, labels.shape[0] - n_images)
    
    imgs, labels = images[random_int : random_int + n_images],\
        labels[random_int:random_int + n_images]
    
    _,figs = plt.subplots(1,n_images,figsize = (n_images *3,3))
    
    for fig, img, label in zip(figs, imgs, labels):
        fig.imshow(img)
        
        ax = fig.axes
        
        ax.set_title(lookup[int(label)])
        
        ax.title.set_fontsize(20)
        
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        
```

```python
# now we will select 5 images at random and display it 
show_img(train_images, train_labels, 5)

# here you can see the 32 * 32 images displayed below 
```

```python
# now we will select 5 images at random and display it 
show_img(train_images, train_labels, 5)

# here you can see the 32 * 32 images displayed below 
```

```python
# before working on the cifar 10 images, we need to store the data set locally 
train_dir = '../data-sets/train/'
test_dir = '../data-sets/test/'

def write_images_to_folders(mode):
    if(mode == 'train'):
        images = train_images
        labels = train_labels
        dir = train_dir
    else:
        images = test_images
        labels = train_labels
        dir = test_dir
    i = 0
    for img,label in zip(images,labels):
        
        path = dir + str(lookup[int(label)])
        
        cv2.imwrite(os.path.join(path, str(i) + '.jpeg'), img)
        
        i+=1
        cv2.waitKey(0)
```

```python
# write the training images and the testing images into the labelled folders
write_images_to_folders('train')
write_images_to_folders('test')
```

```python
# now we have cifar 10 data set on our local machine we can use image data generator 
# in order to read the files and process them. Image data generator allows you
# read the images from any source file or data frame or url. It decodes the images 
# to proper grid or matrix that tensorflow understands and converts them to floating point 
# tensors and performs other processing on images.
# now here the only processsing have specified here is rescaling of the pixel value 
# values of the images so that we divide all pixel value by 255. This will express
# as decimal between o to 1. Ml models generally tend to perform better data which is 
# expressed in small values as fractions rather than large values 
train_image_generator = ImageDataGenerator(rescale = 1./255)

test_image_generator = ImageDataGenerator(rescale = 1./255)
```

```python
# here we will specify the batch size. meaning at a time particualar number of images 
# are taken for training
batch_size = 64
```

```python
# now we are using the train_image_generator.flow_from_directory method 
# in order to read in the images from our local machine. This is for the training set
train_data_gen = train_image_generator.flow_from_directory(batch_size = batch_size,
                                                          directory = train_dir,
                                                          shuffle = True,
                                                          target_size = (32,32))
# in the training data set we see 50k images with 10 classes 
```

```python
# now we are using the test_image_generator.flow_from_directory method 
# in order to read in the images from our local machine. This is for the test set
test_data_gen = test_image_generator.flow_from_directory(batch_size = batch_size,
                                                          directory = test_dir,
                                                          shuffle = True,
                                                          target_size = (32,32))
# in the training data set we see 10k images with 10 classes 
```

```python
# now lets look at the 1 batch of train data invoking the next function on the
# train_data_gen
sample_batch = next(train_data_gen)
sample_batch[0].shape

# here in the first batch we have 64 images with pixels 32 * 32 and with 3 channels rgb
```

```python
# now we will design the cnn architecture using the sequential api's
# we are setting up pf simple sequential model. The sequential model
# consists of stacked convolutional neural networks
conv_model = tf.keras.models.Sequential([
    # first layer is the 2d convolutional. It uses 3*3 convolutional kernal
    # it will generate 16 feature maps. we use the same padding as the output
    # is the input. The activation here we have chosen is relu
    layers.Conv2D(16, (3,3), padding = 'same', activation = 'relu',
                input_shape = sample_batch[0].shape[1:]), 
    layers.Conv2D(32, (3,3), padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(pool_size = (2,2)),
    
    layers.Conv2D(64, (3,3), padding = 'same', activation ='relu'),
    layers.Conv2D(128, (3,3), padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(pool_size = (2,2)),
    
    layers.Flatten(),
    
    layers.Dense(512, activation = 'relu'),
    layers.Dense(256, activation = 'relu'),
    layers.Dense(10, activation = 'softmax')
])
```

```python
# lets now compile the model with its learning paramters 
# the loss function chosen when labels >= 2and when label are one 
# hot encoded (default encoding with the imagedatagenerator)
conv_model.compile(optimizer = 'adam',
                  loss = tf.keras.losses.CategoricalCrossentropy(),
                  metrics = ['accuracy'])
```

```python
# lets now visualise the layers in the network
keras.utils.plot_model(conv_model, 'model_image_classification_shapes.png', show_shapes = True)
```

```python
# now we are ready to start the training the model
training_hist = conv_model.fit(train_data_gen,
                              epochs = 10,
                              steps_per_epoch = len(train_images) // batch_size,
                              validation_data = test_data_gen,
                              validation_steps = len(test_images) // batch_size)
```

```python
# we will use the train history variable to see how the loss and accuracy of the model
# changes during training 
acc = training_hist.history['accuracy']
val_acc = training_hist.history['val_accuracy']

loss = training_hist.history['loss']
val_loss = training_hist.history['val_loss']

epochs_range = range(10)

plt.figure(figsize = (12,8))

plt.subplot(1, 2, 1)

plt.plot(epochs_range, acc, label = 'Training Accuracy')
plt.plot(epochs_range, val_acc, label = 'Validation Accuracy')

plt.legend(loc = 'best')
plt.title('Training and validation Accuracy')
plt.subplot(1, 2, 2)

plt.plot(epochs_range, loss, label = 'Training Loss')
plt.plot(epochs_range, val_loss, label = 'Validation Loss')

plt.legend(loc = 'best')
plt.title('Training and validation Loss')
plt.show()
```

```python
# will set up a helper function called perform test in order to display the 
# prediction results from our models. This takes our model, single image and the 
# corresponding label 
def perform_test(model, img, label):
    plt.imshow(img)
    
    test_img = np.expand_dims(img, axis = 0)
    result = model.predict(test_img)
    
    print('Actual label: ' , lookup[int(label)])
    print('Predicted label: ', lookup[np.argmax(result)])
```

```python
perform_test(conv_model, test_images[0], test_labels[0])
```

```python
perform_test(conv_model, test_images[1], test_labels[1])
```

```python
perform_test(conv_model, test_images[3], test_labels[3])
```

```python
# now are able to see that the the accuracy is more but the val_accuracy is very low. This 
# that we are experiencing an overfitting. lets try to resolve the issue.
# initially we just tweaked just the rescale. now lets retrain on mutiple parameters such as 
# rotation_range, width_shift, height_shift, horizonal_flip and zoom_range
image_gen = ImageDataGenerator(
    rescale = 1. / 255,
    rotation_range = 30,
    width_shift_range = 0.05,
    height_shift_range = 0.05,
    horizontal_flip = True,
    zoom_range = 0.3
)

train_data_gen_aug = image_gen.flow_from_directory(batch_size = batch_size,
                                                  directory = train_dir,
                                                  shuffle = True,
                                                  target_size = (32, 32))

plt.imshow(train_data_gen_aug[0][0][0])
```

```python
# lets chanage the architecture of the neural network as well by adding the dropouts
conv_model_with_dropout = tf.keras.Sequential([
     # first layer is the 2d convolutional. It uses 3*3 convolutional kernal
    # it will generate 16 feature maps. we use the same padding as the output
    # is the input. The activation here we have chosen is relu
    layers.Conv2D(16, (3,3), padding = 'same', activation = 'relu',
                input_shape = sample_batch[0].shape[1:]), 
    layers.Conv2D(32, (3,3), padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(pool_size = (2,2)),
    
    layers.Conv2D(64, (3,3), padding = 'same', activation ='relu'),
    layers.Conv2D(128, (3,3), padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(pool_size = (2,2)),
    
    layers.Dropout(0.10),
    layers.Flatten(),
    
    layers.Dense(512, activation = 'relu'),
    layers.Dropout(0.10),
    layers.Dense(256, activation = 'relu'),
    layers.Dense(10, activation = 'softmax')
])
```

```python
# lets now compile the model with its learning paramters 
# the loss function chosen when labels >= 2and when label are one 
# hot encoded (default encoding with the imagedatagenerator)
conv_model_with_dropout.compile(optimizer = 'adam',
                  loss = tf.keras.losses.CategoricalCrossentropy(),
                  metrics = ['accuracy'])
```

```python
# now we are ready to start the training the model
training_hist = conv_model_with_dropout.fit(train_data_gen,
                              epochs = 100,
                              steps_per_epoch = len(train_images) // batch_size,
                              validation_data = test_data_gen,
                              validation_steps = len(test_images) // batch_size)
```
